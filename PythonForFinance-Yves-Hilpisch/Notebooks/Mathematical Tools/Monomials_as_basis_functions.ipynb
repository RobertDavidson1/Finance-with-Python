## Notation


- $\text{Given a set of so-called basis functions} \quad b_d, d\in\set{1,\dots, D}$
- $\text{We want to find optimal paramters} \quad a_1, \dots a_D$
- $\text{Where} \quad y_i \equiv f(x_i) \quad \text{for} \quad i \in \set{1,\dots, I}$


$$\text{Minimization problem of regression} : \quad\quad \min_{\alpha_1,\ldots,\alpha_D}\frac1I\sum_{i=1}^I\left(y_i-\sum_{d=1}^D\alpha_d\cdot b_d(x_i)\right)^2$$


```python
def f(x):
    return np.sin(x) + 0.5 * x
```

```python
def create_plot(x,y, styles, labels, axlabels):
    plt.figure(figsize=(10,6))
    for i in range(len(x)):
        plt.plot(x[i], y[i], styles[i], label=labels[i])
        plt.xlabel(axlabels[0])
        plt.ylabel(axlabels[1])
    
    plt.legend(loc = 0)
```

```python
x = np.linspace(-2 * np.pi, 2*np.pi, 50)
create_plot([x], [f(x)], ['b'], ['f(x)'], ['x', 'f(x)'])
```

```python
res_deg1 = np.polyfit(x, f(x), deg=1, full=True)
res_deg7 = np.polyfit(x, f(x), deg=5, full=True)
```

```python
ry_deg1 = np.polyval(res_deg1[0],x)
ry_deg7 = np.polyval(res_deg7[0],x)
```

```python
create_plot([x, x, x], [f(x), ry_deg1,  ry_deg7], ['b', 'r.','ko'],
['f(x)','Regression (deg 1)','Regression (deg 7)'], ['x', 'f(x)', 'f(x)'])
```
