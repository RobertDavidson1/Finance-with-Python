```python
import numpy as np
import pandas as pd
from pylab import mpl, plt
import plotly.offline as plyo
plyo.init_notebook_mode(connected=True)
```

Plot styling


```python
style = 'seaborn-v0_8'
plt.style.use(style)
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

Data Import


```python
filename = '../../Data/tr_eikon_eod_data.csv'
data = pd.read_csv(
    filename,
    index_col=0,
    parse_dates=True
    )

df = data[['.SPX','.VIX']].dropna()
df.tail(2)
```

Calculate Logarithimic Returns


```python
rets = np.log(df / df.shift(1))
print(rets.head(3))
rets.dropna(inplace=True)
```

Plot Log Returns of S&P 500 and VIX


```python
rets.plot(subplots=True, figsize=(10,6))
plt.show()
```

```python
pd.plotting.scatter_matrix(rets,
                           alpha=0.2,
                           diagonal='hist',
                           hist_kwds={'bins':35},
                           figsize=(10,6))
plt.show()
```
